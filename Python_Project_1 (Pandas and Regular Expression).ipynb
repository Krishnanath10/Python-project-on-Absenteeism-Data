## Q8: Please convert BMI column into categories. For example, if BMI is < 19.5 code it as under weight, if weight is >= 19.5 and <= 25 code it as normal weight, if weight is >25 and <=30 code it as overweight and >30 code it as Obese.

```python
# Ans

absenteeism_df["BMI in Categories"] = None
for j, i in enumerate(absenteeism_df["Body mass index"]):
    if i < 19.5:
        absenteeism_df["BMI in Categories"][j] = "Under Weight"
    elif i >= 19.5 and i <= 25:
        absenteeism_df["BMI in Categories"][j] = "Normal Weight"
    elif i > 25 and i <= 30:
        absenteeism_df["BMI in Categories"][j] = "Over Weight"
    else:
        absenteeism_df["BMI in Categories"][j] = "Obese"
```

## Q11: Find all words in a string that start with a vowel.
string = 'Apples are red, bananas are yellow, and oranges are orange.'


```python
# Ans

#Importing Regular Expression Library

import re
string = 'Apples are red, bananas are yellow, and oranges are orange.'
pattern = r'\b[aeiouAEIOU]\w*\b'

match_object = re.findall(pattern, string)

for match in (match_object):
    print(match)
```

## Q12: Replace all occurrences of the word "color" with "colour" in a string using regular expressions.

## Q13: Extract all email addresses from a string and print them to the console.
string = 'Please contact us at support@example.com for assistance or info@example.com for more information.'


```python
#Ans

import re

string = 'Please contact us at support@example.com for assistance or info@example.com for more information.'

# defining the pattern
pattern = r'\b[\w.-]+@[\w.-]+\.\w{2,}\b'

# Extracting all the email and print them in new console
match_object = re.findall(pattern, string)

for match in (match_object):
    print(match)
```

## Q14: Validate a US phone number using regular expressions. The phone number should be in the format "XXX-XXX-XXXX" (where X is a digit).
phone_number = '123-456-7890'

```python
# Ans

phone_number = '123-456-7890'

#defining the pattern 
pattern = '^\d{3}-\d{3}-\d{4}$'

# Validating the phone number and print the validation status
new_object = re.match(pattern, phone_number)
if new_object:
    print("The phone number is valid")
else:
    print("The phone numebr is not valid")
```

## Q15: Parse a string of comma-separated key-value pairs and create a dictionary from them.
string = 'key1=value1,key2=value2,key3=value3'

```python
# Ans

# defining the string

string = 'key1=value1,key2=value2,key3=value3'
# defining the pattern
pattern = r'([\w]+)=([\w]+)'

# printing the list of key value pairs
key_value_pairs = re.findall(pattern, string)
print(key_value_pairs)

# creating the dictionary from new list of key value pairs
new_dictionary = {k: v for k, v in key_value_pairs}
print(new_dictionary)
```

## Q16: Convert a string representing a date and time to a datetime object and print it to the console.
date_string = '2023-03-24 16:30:00'

## Q17: Calculate the time difference between two datetime objects in hours and minutes and print it to the console.
start_time = datetime(2023, 3, 24, 16, 30, 0), end_time = datetime(2023, 3, 24, 18, 45, 0)

```python
# Ans
from datetime import datetime

# Define the two datetime objects to calculate the time difference between them
start_time = datetime(2023, 3, 24, 16, 30, 0)
end_time = datetime(2023, 3, 24, 18, 45, 0)

# Calculate the time difference between the two datetime objects
time_difference = end_time - start_time

# Calculate the total number of hours and minutes in the time difference
total_hours = time_difference.seconds // 3600
total_minutes = (time_difference.seconds // 60) % 60

# Print the time difference in hours and minutes to the console
print(f'The time difference between {start_time} and {end_time} is {total_hours} hours and {total_minutes} minutes.')
```
